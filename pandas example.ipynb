## 1. Pandas Series

A **Series** is a one-dimensional labeled array that can hold any data type (integers, strings, floats, etc.). Think of it as a single column of a spreadsheet with an index on the left.

```python
pd.Series([10, 20, 30, 40])
```

## 2. DataFrames

A **DataFrame** is a two-dimensional labeled table — essentially a collection of Series sharing the same index. The most common way to create one is to pass a dictionary where keys become column names and values become lists of rows.

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Salary': [50000, 60000, 70000]
}
df = pd.DataFrame(data)
df
```

## 3. Reading and Writing CSV Files

pandas can persist a DataFrame to disk with `df.to_csv()` and reload it with `pd.read_csv()`. This is the most common file format for tabular data exchange.

```python
df.to_csv("person_output.csv")
```

```python
df = pd.read_csv("person_output.csv")
df
```

## 7. GroupBy and Aggregation

The **split-apply-combine** pattern:
1. **Split** the DataFrame into groups based on a key column (`groupby`)
2. **Apply** an aggregation function (mean, sum, count, …)
3. **Combine** the results into a new DataFrame

Use `.agg()` to apply different functions to different columns simultaneously.

```python
df.groupby("Age")
```

```python
data2 = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Mario'],
    'Age': [25, 30, 35, 25],
    'Salary': [50000, 60000, 70000, 70000]
}
df2 = pd.DataFrame(data2)
df2
```

```python
df2.groupby("Age").agg({
    "Salary": "mean",
    "Name": "count"
})
```

```python
df2.groupby("Age").count()
```

## 8. Merging DataFrames

`pd.merge()` joins two DataFrames like a SQL JOIN:
- `on=` specifies the key column(s) to join on
- `how=` controls the join type: `'left'`, `'right'`, `'inner'`, or `'outer'`

```python
data3 = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Mario', 'Luigi'],
    'Nationality': ['ITA', 'USA', 'USA', 'ITA', 'ITA']
}
df3 = pd.DataFrame(data3)
df3
```

```python
merged_df=pd.merge(df2, df3, on="Name", how="left")
merged_df
```

## 10. Sorting

`df.sort_values(by, ascending=)` sorts rows by one or more columns. Pass a list to `by` for multi-column sorting; `ascending=False` gives descending order.

```python
df.sort_values(['Salary', 'Age'], ascending=False)
```

## 11. Filtering Rows (Boolean Indexing)

A **boolean mask** is a Series of `True`/`False` values with the same index as the DataFrame. Passing it inside `df[mask]` returns only the rows where the condition is `True`.

```python
df2[df2['Age'] < 30]
```

## 12. The `.loc` Accessor

`.loc[row_selector, col_selector]` provides **label-based** selection:
- Filter rows with a boolean condition
- Optionally restrict to specific columns
- Supports **conditional assignment**: `df.loc[condition, 'col'] = value`

```python
df2.loc[df2['Age'] < 30]
```

```python
df2.loc[df2['Age'] < 30, ["Age", "Salary"]]
```

```python
df2.loc[df2['Age'] > 30, "Status"] = "Senior"
```

```python
df2
```

## 13. Applying Functions with `.apply()`

`.apply(func)` passes each element of a Series through `func` and returns a new Series. Use a `lambda` for short one-liners, or a named function for more complex logic.

```python
df2['Age'] = df2['Age'].apply(lambda x: x*0.1)
```

```python
def incr(x: int) -> int:
    return x*0.1
```

```python
df2['Age'].apply(incr)
```
